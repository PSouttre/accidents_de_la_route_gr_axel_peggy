### Suppression des colonnes inutiles


```python
df_lieux.drop(['voie', 'v1', 'v2', 'pr', 'pr1', 'lartpc'], axis=1, inplace=True)
```

```python
df_lieux.head()
```

### Observation des variables

```python
# Catégorie de routes
df_lieux ['catr'].value_counts(normalize = True)
```

```python
# Création de classes
df_lieux['catr'] = df_lieux['catr'].replace(6, 9)
df_lieux['catr'] = df_lieux['catr'].replace(5, 9)
```

```python
# Catégorie de routes
df_lieux['catr'].value_counts(normalize = True)
```

```python
# Régime de circulation
df_lieux['circ'].value_counts(normalize = True)
```

```python
# Suppression des -1
df_lieux = df_lieux[df_lieux['circ']!= -1]
```

```python
# Nombre totale de voie 
df_lieux['nbv'].value_counts(normalize = True)
```

```python
# Suppression de la colonne car manque d'infos métiers 
df_lieux.drop('nbv', axis = 1, inplace = True)
```

```python
# Existence d'une voie réservée
df_lieux['vosp'].value_counts(normalize = True)
```

```python
# Suppression de la colonne car non pertinent
df_lieux.drop('vosp', axis = 1, inplace = True)
```

```python
# Rectiligne/courbe
df_lieux['plan'].value_counts(normalize = True)
```

```python
# Création de classes
df_lieux['plan'] = df_lieux['plan'].replace([3,4], 2)   # En courbe ou en S
```

```python
# Suppression des -1
df_lieux = df_lieux[df_lieux['plan']!= -1]
```

```python
# Rectiligne/courbe
df_lieux['plan'].value_counts(normalize = True)
```

```python
# Largeur chaussée
df_lieux['larrout'].unique()
```

```python
# Suppression de la colonne 'larrout'
df_lieux.drop(['larrout'], axis = 1, inplace = True)
```

```python
# Etat de la surface 
df_lieux['surf'].value_counts(normalize = True)
```

```python
# Création de classes
df_lieux['surf'] = df_lieux['surf'].replace([3,4,6], 2) # Mouillée/flaques/inondées/boue
df_lieux['surf'] = df_lieux['surf'].replace(7, 5)    # Verglas/neige
df_lieux['surf'] = df_lieux['surf'].replace([-1,8], 9)   # Corps gras => autre
```

```python
# Infrastructure
df_lieux['infra'].value_counts(normalize = True)
```

```python
# Suppression de la colonne car non pertinent
df_lieux.drop('infra', axis = 1, inplace = True)
```

```python
# Situation de l'accident
df_lieux['situ'].value_counts(normalize = True)
```

```python
# Suppression de la colonne car non pertinent
df_lieux.drop('situ', axis = 1, inplace = True)
```

```python
# Vitesse max autorisée
df_lieux['vma'].value_counts(normalize = True)
```

```python
# Création de classes
df_lieux['vma'] = df_lieux['vma'].replace([30,20,40,25,45,10,15,35,31,23], 50)
df_lieux['vma'] = df_lieux['vma'].replace([60,65,70,75], 80)
df_lieux['vma'] = df_lieux['vma'].replace(100, 110)
```

```python
# Suppression des -1
df_lieux = df_lieux[df_lieux['vma']!= -1]
```
