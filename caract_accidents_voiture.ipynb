### Suppression des colonnes inutiles


```python
df_caract.drop(['hrmn', 'adr', 'an', 'jour', 'lat', 'long'], axis=1, inplace=True)
```

### Observation des variables

```python
# Mois
df_caract['mois'].value_counts(normalize = True)
```

```python
# Lumière
df_caract['lum'].value_counts(normalize = True)
```

```python
# Création de classes
df_caract['lum'] = df_caract['lum'].replace([3,4,6], 2) # Mouillée/flaques/inondées/boue
df_caract['lum'] = df_caract['lum'].replace(7, 5)    # Verglas/neige
df_caract['lum'] = df_caract['lum'].replace(8, 9)   # Corps gras => autre
```

```python
# Suppression des -1
df_caract = df_caract[df_caract['lum']!= -1]
```

```python
# Département
df_caract['dep'].unique()
```

```python
df_caract['dep'].value_counts(normalize = True)
```

```python
# Suppression de la colonne car non pertinent
df_caract.drop('dep', axis = 1, inplace = True)
```

```python
# Commune
df_caract['com'].value_counts(normalize = True)
```

```python
# Suppression de la colonne car non pertinent
df_caract.drop('com', axis = 1, inplace = True)
```

```python
# Localisation
df_caract['agg'].value_counts(normalize = True)
```

```python
# Intersection
df_caract['int'].value_counts(normalize = True)
```

```python
# Suppression de la colonne car non pertinent
df_caract.drop('int', axis = 1, inplace = True)
```

```python
# Condition atmosphérique
df_caract['atm'].value_counts(normalize = True)
```

```python
# Création de classes
df_caract['atm'] = df_caract['atm'].replace([3,4], 2)  # pluie/neige/grêle
df_caract['atm'] = df_caract['atm'].replace(5, 7)  # pb visibilité
df_caract['atm'] = df_caract['atm'].replace(8, 1)   # normal / couvert
```

```python
# Suppression des -1
df_caract = df_caract[df_caract['atm']!= -1]
```

```python
# type de collision
df_caract['col'].value_counts(normalize = True)
```

```python
# Suppression de la colonne car non pertinent
df_caract.drop('col', axis = 1, inplace = True)
```
