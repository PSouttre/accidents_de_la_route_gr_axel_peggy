Jeu de données pour les véhicules de 93585 lignes pour 11 colonnes.
Un même accident ("Num_Acc") peut concerner plusieurs véhicules ("id_vehicule" et "num_veh") => on peut supprimer "num_veh"

### Suppression des colonnes inutiles


```python
df_vehicules.drop(['num_veh'], axis = 1, inplace = True)
df_vehicules.drop(['occutc'], axis = 1, inplace = True)  #car renseignée qu'à 0.9%
```

```python
df_vehicules.head()
```

### Observation des variables

```python
# Sens de circulation
df_vehicules['senc'].value_counts(normalize = True)
```

```python
# Suppression de la colonne car non pertinente (besoin d'infos métier)
df_vehicules.drop('senc', axis = 1, inplace = True)
```

```python
# Catégorie du véhicule
df_vehicules['catv'].value_counts(normalize = True)
```

```python
# Valeurs de la variable doivent être comprises entre 00 et 99
# Suppression des lignes avec la valeur -1
```

```python
df_vehicules = df_vehicules[df_vehicules['catv'] != -1]
```

```python
# Reduction des classes en 1 seule quand <1%
df_vehicules['catv'] = df_vehicules['catv'].replace([15, 37, 34, 14, 80, 43, 17, 3, 13, 38, 21, 60, 0, 36, 40, 20, 16, 39, 35, 42, 41], 99)
```

```python
# Création de la classe 22 pour les "deux-roues motorisés" => regroupement de 02 – Cyclomoteur <50cm3, 33 – Motocyclette > 125 cm3, 50 – EDP à moteur, 
# 30 – Scooter < 50 cm3, 32 – Scooter > 50 cm3 et <= 125 cm3, 31 – Motocyclette > 50 cm3 et <= 125 cm3
df_vehicules['catv'] = df_vehicules['catv'].replace([2, 33, 50, 30, 31, 32], 22)
```

```python
df_vehicules['catv'].value_counts(normalize = True)
```

```python
# Obstacle fixe heurté
df_vehicules['obs'].value_counts(normalize = True)
```

```python
# Suppression de la colonne car non pertinente (pas d'obstacle dans 88% des cas)
df_vehicules.drop('obs', axis = 1, inplace = True)
```

```python
# Obstacle mobile heurté
df_vehicules['obsm'].value_counts(normalize = True)
```

```python
# Suppression de la colonne car non pertinente 
df_vehicules.drop('obsm', axis = 1, inplace = True)
```

```python
# Point de choc
df_vehicules['choc'].value_counts(normalize = True)
```

```python
# Création de classes
df_vehicules['choc'] = df_vehicules['choc'].replace([2, 3], 1)
df_vehicules['choc'] = df_vehicules['choc'].replace([5, 6], 4)
df_vehicules['choc'] = df_vehicules['choc'].replace(8, 7)
```

```python
# Suppression des -1
df_vehicules = df_vehicules[df_vehicules['choc']!= -1]
```

```python
# Manoeuvre principale avant l'accident 
df_vehicules['manv'].value_counts(normalize = True)
```

```python
# Suppression de la colonne car trop de classe difficile à réunir 
df_vehicules.drop('manv', axis = 1, inplace = True)
```

```python
# Type de motorisation du véhicule
df_vehicules['motor'].value_counts(normalize = True)
```

```python
# Création de classes
df_vehicules['motor'] = df_vehicules['motor'].replace(-1, 0)
df_vehicules['motor'] = df_vehicules['motor'].replace(4, 6)
df_vehicules['motor'] = df_vehicules['motor'].replace(2, 3)
```

## Exportation du dataframe nettoyé

```python
## Exportation du dataframe nettoyé
df_vehicules.to_csv('vehicules_clean.csv', index=False)
```
